> # Showing the domestic gross percentage for each genre at the US market, classified by different production companies

```python
import pandas as pd
```

```python
movies = pd.read_csv('/content/drive/MyDrive/datasets/movies.csv')
movies.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
movies.Studio.unique()
```

```python
new_movies = movies[
    movies.Studio.isin(['Buena Vista Studios', 'Sony', 'Universal', 'WB', 'Paramount Pictures', 'Fox'])
    &
    movies.Genre.isin(['action', 'comedy', 'adventure', 'animation', 'drama'])
]
new_movies
```

```python
sns.set_style('darkgrid')
# sns.set(rc={'figure.figsize':(11.7,8.27)})

f, axes = plt.subplots(figsize=(11.7,8.27))

sns.boxplot(data=new_movies, x='Genre', y='Gross_perc_US', color='lightgrey')
sns.stripplot(data=new_movies, x='Genre', y='Gross_perc_US', hue='Studio', size=8, jitter=True, alpha=0.6)
plt.legend(bbox_to_anchor=(1, 1))

axes.set_title('Domestic Gross % by Genre', fontsize=40)
axes.set_ylabel('Gross % US', fontsize=25)
axes.set_xlabel('Genre', fontsize=25)
```
